```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Read the CSV file
iris_df = pd.read_csv('iris.csv')

# Show first five records
print("First five records:")
print(iris_df.head())

# Show a dataframe which has sepal width greater than 4
print("\nDataframe with sepal width greater than 4:")
sepal_width_gt_4 = iris_df[iris_df['SepalWidthCm'] > 4]
print(sepal_width_gt_4)

# Show a dataframe which has petal width greater than 1
print("\nDataframe with petal width greater than 1:")
petal_width_gt_1 = iris_df[iris_df['PetalWidthCm'] > 1]
print(petal_width_gt_1)

# Retrieve records which have petal width more than 2
print("\nRecords with petal width more than 2:")
petal_width_gt_2 = iris_df[iris_df['PetalWidthCm'] > 2]
print(petal_width_gt_2)

# Scatter plot between sepal length and petal length
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris_df, x='SepalLengthCm', y='PetalLengthCm')
plt.title('Relationship between Sepal Length and Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.grid(True)
plt.show()

# Scatter plot between sepal length and petal length with species as hue
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris_df, x='SepalLengthCm', y='PetalLengthCm', hue='Species')
plt.title('Relationship between Sepal Length and Petal Length with Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.grid(True)
plt.show()
```

Plot 1: The scatter plot shows the relationship between sepal length and petal length. It helps visualize how these two variables are related to each other.


Plot 2: This scatter plot enhances the previous one by adding species as a hue, making it easier to observe how different species vary in terms of sepal length and petal length.


```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

# a. Create dependent variable y and independent variable x
y = iris_df[['SepalLengthCm']]
x = iris_df[['SepalWidthCm']]

# b. Divide the variables into x_train, x_test, y_train, y_test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# c. Show first five records of all four variables
print("x_train:")
print(x_train.head())
print("\nx_test:")
print(x_test.head())
print("\ny_train:")
print(y_train.head())
print("\ny_test:")
print(y_test.head())

# d. Create an object named lr and assign memory from LinearRegression() method
lr = LinearRegression()

# e. Fit both training set into fit method
lr.fit(x_train, y_train)

# f. Predict x_test and store the result into y_pred object
y_pred = lr.predict(x_test)

# h. Show first five records from actual and predicted objects
print("\nFirst five records from actual:")
print(y_test.head())
print("\nFirst five records from predicted:")
print(pd.DataFrame(y_pred, columns=['SepalLengthCm']).head())

# i. Find out mean_squared_error in prediction
mse = mean_squared_error(y_test, y_pred)
print("\nMean Squared Error:", mse)
```
